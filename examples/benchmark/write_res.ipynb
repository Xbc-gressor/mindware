```python
result = dict()
all_model = set()
all_data = set()
```

```python


with open('results.txt', 'r') as file:
    num_epoch = '150'
    for line in file:
        if '--' in line or ':' not in line:
            num_epoch = '101'
            continue
        a  = line.strip()
        scores = a.split(':')[-1]
        model_score = scores.split(',')[0]
        model_ens_score = scores.split(',')[1]
        model_data = a.split(':')[1]
        model = model_data.split(',')[0].split('_')[0] + model_data.split(',')[1] + ' '+num_epoch
        data = model_data.split(',')[-1]
        all_model.add(model)
        all_data.add(data)
        
        if data not in result.keys():   
            result[data] = dict()
        result[data][model] = [model_score.strip(' ')[:7], model_ens_score.strip(' ')[:7]]
```

```python
data_len = len(max(all_data, key=lambda X: len(X)))+2
```

```python
def generate_str(str_, len_):
    str_len = len(str_)
    gap = len_ - str_len
    if gap %2 == 1:
        l1 = int(gap/2)
        l2 = int(gap/2+1)
    else:
        l1 = int(gap/2)
        l2 = int(gap/2)

    return ' '*l1 + str_ + ' '*l2
```

```python
all_model = [' Autogluon one_hot 101',
 ' Autogluon one_hot 150',
 ' Autogluon ord 101',
 ' Autogluon ord 150',
 ' neural one_hot 101',
 ' neural one_hot 150']
```

```python
with open('results_2.txt', 'w') as f:

    f.write('+'+'-' * data_len +('+'+'-'*s_len)*6+'+\n')

    s = '+' + generate_str('Dataset', data_len)
    for m in all_model:
        s += '+' + generate_str(m, s_len)
    s += '+'
    f.write(s+'\n')
    f.write('+'+'-' * data_len +('+'+'-'*s_len)*6+'+\n')

    for data in result.keys():
        s = ''
        s += '+' + generate_str(data, data_len)
        for m in all_model:
            if m in result[data].keys():
                s += '+' + generate_str(result[data][m][0] + ' '+ result[data][m][1], s_len)
            else:
                s += '+' + generate_str(' ',  s_len)
        s += '+'

        f.write(s+'\n')
```
